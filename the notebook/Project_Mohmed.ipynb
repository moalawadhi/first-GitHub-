```python
daily_Tp_values=wq.groupby(wq['Date'])['Tp'].mean()
daily_pH_values=wq.groupby(wq['Date'])['pH'].mean()
daily_Cl_values=wq.groupby(wq['Date'])['Cl'].mean()
daily_Cl_2_values=wq.groupby(wq['Date'])['Cl_2'].mean()
daily_Redox_values=wq.groupby(wq['Date'])['Redox'].mean()
daily_Leit_values=wq.groupby(wq['Date'])['Leit'].mean()
daily_Trueb_values=wq.groupby(wq['Date'])['Trueb'].mean()
daily_Fm_values=wq.groupby(wq['Date'])['Fm'].mean()
daily_Fm_2_values=wq.groupby(wq['Date'])['Fm_2'].mean()
```

```python
Daily_values=pd.DataFrame({'Tp':daily_Tp_values,'pH':daily_pH_values,'Cl':daily_Cl_values,'Cl_2':daily_Cl_2_values,'Redox':daily_Redox_values,'Leit':daily_Leit_values,'Trueb':daily_Trueb_values,'Fm':daily_Fm_values,'Fm_2':daily_Fm_2_values})
```

```python
dv=Daily_values
```

```python
dv=dv.reset_index()
```

Redox potential:reflects the water's ability to break down waste and contaminantes ; and affected mainlyby 3 factors chlorine dioxide ( since chlorine dioxide is a powerful disinfectant and oxidant for pre-oxidation and disinfection used at water treatment), pH and temperature. In general there is a positive relationship between chlorine dioxide and redox potential of water; an increase in chlorine dioxide will leads to an increase in redox potential too.

```python
import matplotlib.pyplot as plt
dv.plot(x='Cl', y='pH', kind='scatter', color='green')
plt.ylabel("pH")
plt.title("Relation Between pH and Chlorine Dioxide Concentration")
plt.xlabel("Chlorine dioxide (mg/L) in line 1");
```

```python
import matplotlib.pyplot as plt
dv.plot(x='Cl_2', y='pH', kind='scatter', color='green')
plt.ylabel("pH")
plt.title("Relation Between pH and Chlorine Dioxide")
plt.xlabel("Chlorine dioxide (mg/L) in line 2");
```

the figure shows kind of negative relationship between  chlorine dioxide concentration at line 2  and pH, as chlorine dioxide concentration increases, the pH tends to decrease. This relationship is chemically plausible. When Chlorine dioxide (ClO2) is dissolved in water, it can react to form various byproducts, some of which are acidic. For example, the formation of chlorite(ClO 2-) and other species can release protons (H+) into the solution, which lowers the pH. Same relation with respect to  chlorine dioxide concentration at line 1 (but much clearer at line 2).

```python
import matplotlib.pyplot as plt
dv.plot(x='Fm_2', y='Cl_2', kind='scatter', color='red')
plt.ylabel("Chlorine Dioxide Concentration (line 2) (mg/L)")
plt.title ("Relation Between Chlorine Dioxide Concentration and Water Flow rate ")
plt.xlabel("Water Flow rate (line_2) (m3/h)");
```

Generally,the figure shows a positive correlation between Chlorine Dioxide concentration and water flow rate at line 2,  as the water flow rate increases, the Chlorine Dioxide concentration also tends to increase. This relationship is plausible in a water treatment or distribution system, an increase in the flow rate must be followed by an increase in the chlorine dioxide concentration to maintains a target concentration.

```python
import matplotlib.pyplot as plt
dv.plot(x='Fm', y='Trueb', kind='scatter', color='red')
plt.ylabel("Truebidty (NTU)")
plt.title ("Relation Between Water Flowrate and Truebidty")
plt.xlabel("Water Flowrate (line 1) (m3/h)");
```

The figure shows almost a positive reationship between water flow rate at line 1 and trubdity of water, as flowrate increases, truebdity inceased too. Which means that an increase in flowrate probably increased the suspended particles and dissolved substances, thus an increase in trubdity showed up.
